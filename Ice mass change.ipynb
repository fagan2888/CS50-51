```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

filename = "https://course-resources.minerva.kgi.edu/uploaded_files/mke/rx9A0n/s1-student-version--ea-assignment--climate-change--land-ice-mass-data---sheet1.csv"
data = pd.read_csv(filename, skiprows=9, header=1)
data.head(5)

```

```python
plt.plot(data['TIME (year.decimal fraction)'], data['Greenland mass (Gt)'])
plt.plot(data['TIME (year.decimal fraction)'], data['Antarctica mass (Gt)'])
plt.legend(["Greenland", "Antarctica"])
plt.title("Change in polar ice masses")
plt.xlabel("Year")
plt.ylabel("Annual land ice mass anomalies from mean (Gt)")
```

```python
plt.scatter(data['TIME (year.decimal fraction)'], data['Greenland mass (Gt)'])
plt.scatter(data['TIME (year.decimal fraction)'], data['Antarctica mass (Gt)'])
plt.legend(["Greenland", "Antarctica"])
plt.title("Change in polar ice masses")
plt.xlabel("Year")
plt.ylabel("Annual land ice mass anomalies from mean (Gt)")
```

```python
plt.hist(data['Greenland mass (Gt)'], fc=(0, 0, 1, 0.7))
plt.hist(data['Antarctica mass (Gt)'], fc=(1, 0, 0, 0.7))
plt.legend(["Greenland", "Antarctica"])
plt.title("Distribution of polar ice mass anomalies")
plt.xlabel("Annual land ice mass anomalies from mean (Gt)")
plt.ylabel("Number of occurences")
```
